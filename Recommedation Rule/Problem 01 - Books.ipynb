#### 4. Data Preparation

```python
book_corr_reco = pd.pivot_table(data = book_data_1,values='Book.Rating',index='User.ID',columns='Book.Title').fillna(0)
book_corr_reco
```

```python
book_corr_reco.index
```

#### Using Cosine distance as a Metric for UBCF

```python
cosine_df = pd.DataFrame(sklearn.metrics.pairwise_distances(X = book_corr_reco,metric='cosine')).round(4)
cosine_df
```

```python
np.fill_diagonal(a = cosine_df.to_numpy(),val = 1)
cosine_df
```

```python
cosine_df.index = book_corr_reco.index
```

```python
cosine_df.columns= book_corr_reco.index
```

```python
cosine_df
```

```python
df = pd.DataFrame(cosine_df.idxmin())
df.columns = ['corr']
df
```

```python
Reco_df = df[df['corr'] != 8]
Reco_df
```
