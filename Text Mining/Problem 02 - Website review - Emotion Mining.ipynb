#### 2. Importing the data

```python
url = 'https://www.flipkart.com/boat-airdopes-131-upto-15-hours-playback-13mm-drivers-iwp-technology-bluetooth-headset/product-reviews/itmf77ad6e8eecf1?pid=ACCFSDGX68PUKJYE&lid=LSTACCFSDGX68PUKJYEHF5WGJ&aid=overall&certifiedBuyer=false&sortOrder=MOST_RECENT'
```

```python
page = requests.get(url)
page
```

```python
page.text
```

```python
soup = bs(page.text,'html.parser')
print(soup.prettify())
```

```python
reviews = soup.find_all( 'div',class_="t-ZTKy")
reviews
```

```python
review = []
for item in reviews:
    item = item.get_text().replace('\n','')
    item = item.strip()
    review.append(item)
```

```python
review
```

```python
new_review  = []
for k in review:
    k = k.replace('READ MORE', '')
    new_review.append(k)
```

```python
new_review
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
sentiment_pipeline(new_review)
```
